```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load dataset

athlete = pd.read_csv('athlete_events.csv')
region = pd.read_csv('noc_regions.csv')
```

```python
athlete.head()
```

```python
region.head()
```

```python
# Join the dataframes

athlete_df = athlete.merge(region, how = 'left', on = 'NOC')
athlete_df.head()
```

```python
athlete_df.shape
```

```python
# Columns name consistent

athlete_df.rename(columns= {'region': 'Region', 'notes': 'Notes'}, inplace=True)
```

```python
athlete_df.head()
```

```python
athlete_df.info()
```

```python
athlete_df.describe()
```

```python
# Check null values

nan_values = athlete_df.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
athlete_df.isnull().sum()
```

```python
# India details

athlete_df.query('Team == "India"').head(5)
```

```python
# Japan details

athlete_df.query('Team == "Japan"').head(5)
```

```python
# Top countries participating

top_10_countries = athlete_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Plot for the top 10 countries

plt.figure(figsize=(12,6))
plt.xticks(rotation=20)
plt.title('Overall Participation of Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette = 'Set2')
```

```python
# Age Distribution of the participants

plt.figure(figsize=(12,6))
plt.title("Age Distribution of the Athletes")
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athlete_df.Age, bins=np.arange(10,80,2), color='orange', edgecolor='white')
plt.show()
```

```python
# Winter Olympics Sports

winter_sports = athlete_df[athlete_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
#Summer Olympics Sports

summer_sports = athlete_df[athlete_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
# Male and Female participants

gender_counts = athlete_df.Sex.value_counts()
gender_counts
```

```python
# Pie Plot for male and female athletes

plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=180, shadow=True)
plt.show()
```

```python
# Total Medals

athlete_df.Medal.value_counts()
```

```python
# Total number of female athletes in each olympics

female_participants = athlete_df[(athlete_df.Sex == 'F') & (athlete_df.Season == 'Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
womenOlympics = athlete_df[(athlete_df.Sex == 'F') & (athlete_df.Season == 'Summer')]
womenOlympics.head()
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womenOlympics, palette='Spectral')
plt.title('Women Participation')
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes over time')
```

```python
# Gold medal athletes

goldMedals = athlete_df[(athlete_df.Medal == 'Gold')]
goldMedals.head()
```

```python
# Take only the valuesthat are different from NaN.

goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
# Gold beyond 60

goldMedals['ID'][goldMedals['Age'] > 60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age'] > 60]
sporting_event
```

```python
# Gold medals from each country

goldMedals.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x="Region", y="Medal", data=totalGoldMedals, height=5, kind='bar', palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 countries')
g.set_ylabels('Number of Medals')
plt.title('Gold Medals per country')
```

```python
# Rio olympics

max_year = athlete_df.Year.max()
print(max_year)

team_names = athlete_df[(athlete_df.Year == max_year) & (athlete_df.Medal == 'Gold')].Team

team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel('Countrywise Medals for the year 2016')
```
